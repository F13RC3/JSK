```python
import pandas as pd
import numpy as np
import regex as rx
import difflib
```

```python
name2019 = pd.read_excel('E:/maharshtra/khatam.qgz_FILES/AC_072_LS 2019_Form 201.xlsx')
vs2014 = pd.read_excel('E:/maharshtra/all Files/2014 Form 20(copy).xlsx')
```

```python

# uni_vil.head
# name2019['PB Name']
# vs2014['PB Name(2014)'][1:]
```

```python
vill_name = name2019['PB Name']
vilN = vs2014['PB Name(2014)'][1:]
```

```python
sim_names = []
for i in range(len(vill_name)):
    a = difflib.get_close_matches(vill_name[i], vilN, cutoff=0.7)
    if not a:
        a = ['']
    sim_names.append([a[0], vill_name[i]])
```
