# This part involves Converting pdfs 1st page to JPEG and saving it 

```python
def split(a, n):
    k, m = divmod(len(a), n)
    return (a[i * k + min(i, m):(i + 1) * k + min(i + 1, m)] for i in range(n))
#Splitting files into 500 parts to increase frequency of saving. Increase the value to save more frequently
#Do not change value after running code once. If want to change value after running. Delete all saved files or do maths
```

```python
#getting all pdfs as well as dividing it in 13 parts using split function
pdfs = glob.glob('/f_drive/Pdf of Constituencies/ballarpur/*.pdf')
pdf_fin_split =  list(split(pdfs, 13))
```

```python
#Extracting only 1st page of pdfs as jpeg
count = 1
for i in pdf_fin_split[count-1:]:
    for j in i:
        pages = convert_from_path(j, last_page=1)
        pages[0].save('/f_drive/Pdf of Constituencies/images ballarpur/'+j.split('/')[-1][:-4]+'.jpg', 'JPEG')
    count+=1
```

# Main Part for Hindi Text Extraction 

```python
from PIL import Image
from pytesseract import image_to_string
import pytesseract
import numpy as np
import pandas as pd
import PIL
# from PIL import Image, ImageFilter, ImageEnhance
import cv2
```

```python
page_1 = glob.glob('/f_drive/Pdf of Constituencies/images ballarpur/*.jpg')
page_fin_split = list(split(page_1, 13))
# page_fin_split
```

```python
def img_to_txt(im):
    crop_rectangle_vs = (710, 783, 1486,  928)
    cropped_im_vs = im.crop(crop_rectangle_vs)
    
    v_details = list(filter(str.strip, image_to_string(cropped_im_vs, lang='hin').replace("4","").split('\n')))
    vill_name = v_details[0][19:]
    police_stat = v_details[1][12:]
    taluka = v_details[2][7:]
    jilha = v_details[3][8:]
    
#     crop_rectangle_bs = (1462, 110, 1550, 205)
#     cropped_im_bs = im.crop(crop_rectangle_bs)
    crop_rectangle_pin = (1091, 928, 1486, 970)
    cropped_im_pin = im.crop(crop_rectangle_pin)
    pin_code = image_to_string(cropped_im_pin, config='-c tessedit_char_whitelist=0123456789 --psm 13 ')
    
    crop_rectangle_psn = (76, 1347, 861, 1404)
    cropped_im_psn = im.crop(crop_rectangle_psn)
    stat_name = image_to_string(cropped_im_psn, lang='hin')
    
    crop_rectangle_psa = (76, 1450, 861, 1517)
    cropped_im_psa = im.crop(crop_rectangle_psa)
    stat_add = image_to_string(cropped_im_psa, lang='hin')
    
    crop_rectangle_male = (554, 1687, 827, 1718)
    cropped_im_male = im.crop(crop_rectangle_male)
    male = image_to_string(cropped_im_male, config='-c tessedit_char_whitelist=0123456789 --psm 13 ')
    
    crop_rectangle_fem = (836, 1688, 1014, 1718)
    cropped_im_fem = im.crop(crop_rectangle_fem)
    female = image_to_string(cropped_im_fem, config='-c tessedit_char_whitelist=0123456789 --psm 13 ')
    
    crop_rectangle_tot = (1276, 1690, 1508, 1718)
    cropped_im_tot = im.crop(crop_rectangle_tot)
    tot = image_to_string(cropped_im_tot, config='-c tessedit_char_whitelist=0123456789 --psm 13 ')
    
    cropped_im_oth = im.crop(crop_rectangle_oth)
    crop_rectangle_tot = (1276, 1690, 1508, 1718)
    oth = image_to_string(cropped_im_oth, config='-c tessedit_char_whitelist=0123456789 --psm 13 ')
    
    crop_rectangle_fsn = (77, 1690, 258, 1722)
    cropped_im_fsn = im.crop(crop_rectangle_fsn)
    fsn = image_to_string(cropped_im_fsn, config='-c tessedit_char_whitelist=0123456789 --psm 13 ')
        
    crop_rectangle_lsn = (266, 1690, 548, 1722)
    cropped_im_lsn = im.crop(crop_rectangle_lsn)
    lsn = image_to_string(cropped_im_lsn, config='-c tessedit_char_whitelist=0123456789 --psm 13 ')
    
    crop_rectangle_ser = (1280, 159, 1509, 207)
    cropped_im_ser = im.crop(crop_rectangle_ser)
    serial = image_to_string(cropped_im_ser, config='-c tessedit_char_whitelist=0123456789 --psm 13 ')
    
    return serial,vill_name, police_stat, taluka, jilha, pin_code, stat_name, stat_add, male, female, oth, tot, fsn, lsn
```

```python
count = 1
vill_name = []
police_station = []
taluka = []
jilha = []
pincode = []
male = []
female = []
total = []
polling_station = []
stat_add = []
booth_name = []
fsnl = []
lsnl = []
others = []
# serial = 1
s_n = []

for i in page_fin_split[count-1:]:


    for j in i:
#         print(j)
        x = j
        im = Image.open(x)
        im = cv2.bilateralFilter(np.array(im, dtype= np.uint8),9,75,75)
        im = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
        im = cv2.adaptiveThreshold(np.array(im,dtype=np.uint8), 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 31, 15)
        im = Image.fromarray(im)
        
        serial,v,p,ta,j,pin,sn,sa,m,f,oth,tot,fsn,lsn = img_to_txt(im)
        
        cropped_im_n = im.crop((76, 754, 696, 1247))
        # cropped_im_n
        booth = list(filter(str.strip, image_to_string(cropped_im_n, lang='hin').split('\n')))
        
#         for k in booth:
        s_n.append(serial)
        male.append(m)
        female.append(f)   
        others.append(oth)
        total.append(tot)
        vill_name.append(v)
        police_station.append(p)
        taluka.append(ta)
        jilha.append(j)
        pincode.append(pin)                
        polling_station.append(sn)
        stat_add.append(sa)
        booth_name.append(', '.join(booth))    
        fsnl.append(fsn)
        lsnl.append(lsn)
            
    print(count)
    count+=1
#     break
```

```python
# Saving all the data into a single excel file


df = pd.DataFrame(np.column_stack([s_n, polling_station, stat_add, vill_name, 
                                       police_station, taluka, jilha, pincode, booth_name, fsnl, lsnl, male, female, others,
                                      total]), 
                   columns=['Station No' ,'Polling Station','Station Address', 'Village Name', 
                            'Police Station', 'Taluka','Jilha','Pincode','Booth Name', 'First Serial No', 'Last Serial No', 'Male', 'Female', 'Others', 'Total'])
df.to_excel("/f_drive/Pdf of Constituencies/FirstPageData_no_reduntant.xlsx",index = False)
```
