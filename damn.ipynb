```python
ac = []
pn = []
sn = []
name = []
eid = []
gen = []
age = []
for i in range(len(alphas)):
    try:
        df = pd.read_excel('/f_drive/Pdf of Constituencies/maharashtra (ballarpur)/'+alphas[i]+"Maharashtra(ballarpur).xlsx")
        ac.extend(df.iloc[:,0])
        pn.extend(df.iloc[:,1])
        sn.extend(df.iloc[:,2])
        name.extend(df.iloc[:,3])
        eid.extend(df.iloc[:,4])
        gen.extend(df.iloc[:,5])
        age.extend(df.iloc[:,6])
    except:
        continue
```

```python
df = pd.DataFrame(np.column_stack([ac, pn, sn, name, eid, gen, age]), 
                           columns=['Assembly Constituency','Part No', 'Serial No', 'Name', 'Epic ID','Geder','Age'])
df.to_excel("/f_drive/Pdf of Constituencies/maharashtra (ballarpur)/"+"Complete_Maharashtra(ballarpur).xlsx",index = False)
```
